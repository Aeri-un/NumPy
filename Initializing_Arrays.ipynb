```python
import numpy as np
```

```python
# All 0s matrix
print(np.zeros(5))
print(np.zeros((2,2)))
print(np.zeros((2,3,3)))
```

```python
# All 1s matrix
print(np.ones((3,2,2)))
print(np.ones(((3,2,2)), dtype='int16'))
```

```python
# Any other number 
print(np.full((2,2),99,))
print(np.full((2,2),99, dtype='float32'))
```

```python
# Any other number (full-like)
a = np.array(([1,2,3,4],[5,6,7,8]))
print(a)
print(np.full_like(a,4))
print(np.full(a.shape,4))
```

```python
# Random decimal number 
print(np.random.rand(4,2))
print(np.random.random_sample(a.shape))
```

```python
# Random integer values 
print(np.random.randint(7))
print(np.random.randint(7,size=(3,3)))
print(np.random.randint(4,7,size=(3,3))) 
# 7 is excluded
```

# Identity matrix
print(np.identity(3))
print(np.identity(5,dtype='int16'))

```python
# Repeat an array
arr = np.array([1,2,3])
r1 = np.repeat(arr,3)
print(r1)

arr1 = np.array([[1,2,3]])
r2=np.repeat(arr1,3)
r3=np.repeat(arr1,3,axis=0)
print(r2)
print(r3)
```

```python
# Prac
b = np.ones((5,5),dtype='int16')
print(b)
z = np.zeros((3,3))
z[1,1] = 9
print(z)
b[1:-1,1:-1]=z
# b[1:4,1:4]=z
print(b)
```

```python
# Be careful when copying arrays 
a = np.array([1,2,3])
b = a
print(a)
print(b)

b[0] = 100
print(b)
print(a)

# Same like in java b is not a copy of a rather it is pointing to a
```

```python
# Instead use .copy() function
a = np.array([1,2,3])
b = a.copy()
print(a)
print(b)

b[0] = 100
print(b)
print(a)
```
