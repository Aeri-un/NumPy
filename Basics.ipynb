```python
import numpy as np
```

```python
# BASICS
a = np.array([1,2,3])
print(a)
a
```

```python
b = np.array([[9.0,8.0,7.0],[1.0,2.0,3.0]])
print(b)
b
```

```python
# Get dimension - Gives number of dimensions 
print(a.ndim)
print(b.ndim)
```

```python
# Get shape - Gives exact dimentions 
print(a.shape)
print(b.shape)
```

```python
# Get type
print(a.dtype)

#Specify data type which we can to use
c = np.array([1,2,3,4], dtype='int16')
print(c.dtype)
```

```python
# Get Size - Size in bytes 
print(a.itemsize)
print(c.itemsize)
print(b.itemsize)
```

```python
# Get total size - Total size of array in bytes 
print(a.size*a.itemsize)
print(c.size*c.itemsize)
print(b.size*b.itemsize)

# Or directly use 
print(a.nbytes)
print(c.nbytes)
print(b.nbytes)
```
