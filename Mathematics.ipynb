```python
import numpy as np
```

```python
a = np.array([1,2,3,4])
print(a)
print(a+2)
print(a/2)
a-=2
print(a)
```

```python
a = np.array([1,2,3,4])
b = np.array([1,0,1,0])
print(a+b)
print(a**2)
```

```python
# Take the sin of all values 
np.sin(a)
```

```python
 # Linear Algebra 

a = np.ones((2,3))
print(a)

b = np.full((3,2),2)
print(b)

print(np.matmul(a,b))
```

```python
# Determinant of a matrix 
c = np.identity(3)
print(np.linalg.det(c)) 
```
